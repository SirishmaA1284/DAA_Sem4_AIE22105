```python
#Q1
import time
import matplotlib.pyplot as plt

# Iterative approach to find the sum of first N natural numbers
def iterative_sum(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i
    return sum

# Recursive approach to find the sum of first N natural numbers
def recursive_sum(n):
    if n == 0:
        return 0
    else:
        return n + recursive_sum(n - 1)

# Function to measure execution time of given function
def measure_time(func, *args):
    start_time = time.time()
    func(*args)
    end_time = time.time()
    return end_time - start_time

# Varying N's value and plotting time taken
n_values = list(range(1, 1001, 100))
iterative_times = []
recursive_times = []

for n in n_values:
    iterative_times.append(measure_time(iterative_sum, n))
    recursive_times.append(measure_time(recursive_sum, n))

plt.plot(n_values, iterative_times, label='Iterative')
plt.plot(n_values, recursive_times, label='Recursive')
plt.xlabel('Value of N')
plt.ylabel('Time taken (seconds)')
plt.title('Comparison of Iterative and Recursive Sum Calculation')
plt.legend()
plt.show()
```

```python
#Q2
import time
import random
import matplotlib.pyplot as plt

# Linear search algorithm
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Binary search algorithm
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generating array of 10000 random integers
array = [random.randint(1, 1000) for _ in range(10000)]

# Input from user for search key
search_key = int(input("Enter the search key: "))

# Measuring time for linear search
linear_times = []
for _ in range(5):
    start_time = time.time()
    linear_search(array, search_key)
    end_time = time.time()
    linear_times.append(end_time - start_time)

# Measuring time for binary search
binary_times = []
for _ in range(5):
    start_time = time.time()
    binary_search(array, search_key)
    end_time = time.time()
    binary_times.append(end_time - start_time)

# Plotting the time taken by each algorithm
plt.plot(range(1, 6), linear_times, label='Linear Search')
plt.plot(range(1, 6), binary_times, label='Binary Search')
plt.xlabel('Search Attempt')
plt.ylabel('Time taken (seconds)')
plt.title('Comparison of Linear and Binary Search Time')
plt.legend()
plt.show()
```

```python
#Q3
def string_to_int(string):
    if len(string) == 1:
        return int(string)
    else:
        return int(string[0]) * 10**(len(string)-1) + string_to_int(string[1:])

string = input("Enter a string of digits: ")
integer = string_to_int(string)
print("Integer representation:", integer)
```

```python
#Q4
def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

s = "pots&pans"
reversed_s = reverse_string(s)
print("Reversed string:", reversed_s)
```

```python
#Q5
def is_palindrome(s):
    if len(s) <= 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return is_palindrome(s[1:-1])

string = input("Enter a string: ")
if is_palindrome(string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```
